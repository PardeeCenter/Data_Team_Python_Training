```python
# very explicit but not efficient version
value=dt2.Country[0]
value_list=value.split(",")
new_value_list=[]
for i in value_list:
    i_new=i.strip()
    new_value_list.append(i_new)
#
new_value_list
```

```python
# or you could just
[i.strip() for i in dt2.Country[0].split(",")]
```

```python
year=dt2['Year']
country=dt2['Country']
drop_index=[]
for i in range(dt2.shape[0]):
    if ',' not in str(year[i]) and ',' not in country[i]:
        continue
    else:
        drop_index.append(i)
        y_split=[j.strip() for j in str(year[i]).split(',')]
        country_split=[j.strip() for j in country[i].split(',')]
        dt_row=dt2.iloc[i,:].copy(deep=True)
        
        for y in y_split:
            for c in country_split:
                dt_row['Country']=c
                dt_row['Year']=int(y)
                dt2=dt2.append(dt_row,ignore_index=True,sort=False)
```

```python
dt2
```

```python
dt2=dt2.drop(index=drop_index)
```
