Question: How do you know if a string is a part of a line of text?

```python
txt1="Yeah! Apple just launched the brand new iPhone 11. But I don't care, I'm not able to afford it."
txt2="launch"
len_1=len(txt1)
len_2=len(txt2)
```

```python
for i in range(0,len_1-len_2):
    sub_txt1=txt1[i:i+len_2]
    
    # test if two strings are exact the same
    compare=0
    for j in range(0,len_2):
        if sub_txt1[j]==txt2[j]:
            compare+=1
    # outout the result and starting position of the match
    if compare==len_2:
        print("True",i)
```

# Regular Expression 

https://docs.python.org/3/library/re.html

Question: What if I want the whole sentence? - which company launched what? <br>

\[ \] - matches a set of characters <br>
\[a-z\] - matches a any characters from a to z <br>
\[A-Z\] - matches a any characters from A to Z <br>
\s - matches whitespace <br>
\+ - causes the resulting RE to match 1 or more repetitions of the preceding RE



```python
"Yeah! Apple just launched the brand new iPhone 11. But I don't care, I'm not able to afford it."
```

```python
re.findall(r"[a-zA-Z\s]+launched[a-zA-Z\s0-9]+",txt1)[0].strip()
```

- re.search(pattern, string, flags=0) <br>
Scan through string looking for the first location where the regular expression pattern produces a match, and return a corresponding match object. Return None if no position in the string matches the pattern; note that this is different from finding a zero-length match at some point in the string.

```python
re.search("[^a-zA-Z]+","Côte d'Ivoire")
```

```python
"Côte d’Ivoire"[1:2]
```

- re.findall <br>
Return all non-overlapping matches of pattern in string, as a list of strings. The string is scanned left-to-right, and matches are returned in the order found. If one or more groups are present in the pattern, return a list of groups; this will be a list of tuples if the pattern has more than one group. Empty matches are included in the result.
Changed in version 3.7: Non-empty matches can now start just after a previous empty match.

```python
re.findall("[^a-zA-Z]","Côte d’Ivoire")
```
