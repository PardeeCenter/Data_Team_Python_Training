"Python is an â€œobject-oriented programming language. <br>This means that almost all the code is implemented using a special construct called classes"<br>

```python
class PardeeMember:
    # Attributes
    def __init__(self,name,gender):
        self.name=name
        self.gender=gender
        
        if gender=="male":
            self.gender2="he"
        elif gender=="female":
            self.gender2="she"
        
    # Methods
    def isboss(self):
        if self.name=="Yutang":
            return(f"Right, {self.gender2} is")
        else:
            return(f"Sorry, {self.gender2} is not")
```
