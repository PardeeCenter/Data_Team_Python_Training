# Read in data

```python
dt=pd.read_excel('Sample Data/Test1Data.xlsx',sheet_name="Data")
dt_region=pd.read_excel('Sample Data/Test1Data.xlsx',sheet_name='Country Region')
```

# Add in regions

```python
dt=pd.merge(left=dt,right=dt_region,on=["country_a"],how="left")
dt_region.rename(columns={"country_a":"country_b","region_a":"region_b","subregion_a":"subregion_b"},inplace=True)
dt=pd.merge(left=dt,right=dt_region,on=["country_b"],how="left")
```

# Upper Lower Case

```python
dt["country_a"]=dt["country_a"].apply(lambda x: x.title())
dt["country_b"]=dt["country_b"].apply(lambda x: x.title())
dt["country_a"]=dt["country_a"].apply(lambda x: re.sub(r" And ",r" and ",x))
dt["country_b"]=dt["country_b"].apply(lambda x: re.sub(r" And ",r" and ",x))
```

```python
dt.head()
```

# Remove duplicates & fill in nas

```python
dt.shape
```

```python
print(dt.shape[0])
dt.drop_duplicates(inplace=True)
print(dt.shape[0])
```

```python
dt.fillna(0,inplace=True)
```

# Sort the data and get the moving average

```python
dt.sort_values(by=["country_a","country_b","year"],inplace=True)
```

```python
dt_avg=dt.groupby(["country_a","country_b"])['ln_totaltrade'].rolling(3,min_periods=1).mean().reset_index()
dt_avg.head(20)
dt["ln_totaltrade_ma"]=dt_avg["ln_totaltrade"]
```
